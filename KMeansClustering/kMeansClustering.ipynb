```python
import numpy as np
from math import sqrt
import matplotlib.pyplot as mplt
import matplotlib.markers as mmarkers

def plotClassificationGraph(currMus, currA, K, X, N, D, fileName):
	fig, ax = mplt.subplots()
	X_x = X[:,0]
	X_y = X[:,1]
	temp1 = []
	i=0
	while(i<N):
		j=0
		while(j<K):
			if(currA[i][j]==1):
				temp1.append(j+1)
			j=j+1
		i=i+1
	temp1 = np.array(temp1)
	ax.scatter(X_x[temp1==1],X_y[temp1==1], c='red', facecolors='full', marker='^', edgecolors='blue')
	ax.scatter(X_x[temp1==2],X_y[temp1==2], c='green', facecolors='full', marker='^', edgecolors='blue')
	ax.scatter(X_x[temp1==3],X_y[temp1==3], c='blue', facecolors='full', marker='^', edgecolors='blue')
	i=0
	while(i<N):
		txt = "("+((str)(X_x[i]))+","+((str)(X_y[i]))+")"
		ax.annotate(txt, (X_x[i], X_y[i]))
		i=i+1
	Mu_x = np.around((currMus[:,0]).copy(), decimals=3)
	Mu_y = np.around((currMus[:,1]).copy(), decimals=3)
	temp2 = np.array([1, 2, 3])
	ax.scatter(Mu_x[temp2==1],Mu_y[temp2==1], c='red', facecolors='full', marker='o', edgecolors='red')
	ax.scatter(Mu_x[temp2==2],Mu_y[temp2==2], c='green', facecolors='full', marker='o', edgecolors='green')
	ax.scatter(Mu_x[temp2==3],Mu_y[temp2==3], c='blue', facecolors='full', marker='o', edgecolors='blue')
	i=0
	while(i<K):
		txt = "("+((str)(Mu_x[i]))+","+((str)(Mu_y[i]))+")"
		ax.annotate(txt, (Mu_x[i], Mu_y[i]))
		i=i+1
	fig.savefig(fileName, dpi=fig.dpi)
	
def plotMuGraph(nextMus, currA, K, X, N, D, fileName):
	fig, ax = mplt.subplots()
	X_x = X[:,0]
	X_y = X[:,1]
	ax.scatter(X_x, X_y, facecolors='none', marker='^', edgecolors='blue')
	i=0
	while(i<N):
		txt = "("+((str)(X_x[i]))+","+((str)(X_y[i]))+")"
		ax.annotate(txt, (X_x[i], X_y[i]))
		i=i+1
	Mu_x = np.around((nextMus[:,0]).copy(), decimals=3)
	Mu_y = np.around((nextMus[:,1]).copy(), decimals=3)
	temp2 = np.array([1, 2, 3])
	ax.scatter(Mu_x[temp2==1],Mu_y[temp2==1], c='red', facecolors='full', marker='o', edgecolors='red')
	ax.scatter(Mu_x[temp2==2],Mu_y[temp2==2], c='green', facecolors='full', marker='o', edgecolors='green')
	ax.scatter(Mu_x[temp2==3],Mu_y[temp2==3], c='blue', facecolors='full', marker='o', edgecolors='blue')
	i=0
	while(i<K):
		txt = "("+((str)(Mu_x[i]))+","+((str)(Mu_y[i]))+")"
		ax.annotate(txt, (Mu_x[i], Mu_y[i]))
		i=i+1
	fig.savefig(fileName, dpi=fig.dpi)

def printOP(X, currA, currMus, N, K):
	txt = ""
	l=0
	while(l<K):
		txt1 = ""
		m=0
		while(m<N):
			if (currA[m][l]==1):
				txt1 = txt1 + str(X[m]) + " , "
			m=m+1
		txt = txt + str(l+1)+". "+str(currMus[l])+" - "+txt1[:len(txt1)-3] + "\n"
		l=l+1
	print(txt)

class KMeansClustering():

	def __init__(self):
		self.Mus = None
		self.A = None
	
	def fit(self, K, X, initialMusGiven = False, initialMus=None):
		N = len(X) # Number of rows in X
		D = len(X[0]) # Number of cols in X. Also is the dimension of X
		I=0
		while(1==1):
			# Step 1 - Expectation : Find optimal A
			if (I==0):
				if (initialMusGiven==True):
					currMus = initialMus
				else:
					currMus = X[0:K]
			else:
				currMus = nextMus
			currA = np.zeros((N, K))
			i=0
			while(i<N):
				x = X[i]
				minEuclidean_j = 0
				minEuclidean_val = 1000000000
				j=0
				while(j<K):
					ED = sqrt(np.sum(np.square(currMus[j]-x)))
					if ( ED < minEuclidean_val ):
						minEuclidean_val = ED
						minEuclidean_j = j
					j=j+1
				currA[i][minEuclidean_j] = 1 
				i=i+1
			if (I==0 and D==2):
				plotClassificationGraph(currMus, currA, K, X, N, D, 'task3_iter1_a.jpg')
				print("3.1-")
				printOP(X, currA, currMus, N, K)
			if (I==1 and D==2):
				plotClassificationGraph(currMus, currA, K, X, N, D, 'task3_iter2_a.jpg')
				print("3.3-")
				printOP(X, currA, currMus, N, K)
			# Step 2 - Maximization : Find optimal Mus
			nextMus = []
			j=0
			while(j<K):
				sm = np.zeros(D)
				cnt = 0
				i=0
				while(i<N):
					if (currA[i][j]==1):
						cnt = cnt + 1
						sm = sm + X[i]
					i=i+1
				mean = (sm/cnt)
				nextMus.append(mean)
				j=j+1
			nextMus = np.array(nextMus)
			if (I==0 and D==2):
				plotMuGraph(nextMus, currA, K, X, N, D, 'task3_iter1_b.jpg')
				print("3.2-")
				print(nextMus)
				print()
			if (I==1 and D==2):
				plotMuGraph(nextMus, currA, K, X, N, D, 'task3_iter2_b.jpg')
				print(nextMus)
			I=I+1
			if (np.array_equal(currMus, nextMus)==True):
				break

		self.A = currA
		self.Mus = currMus
```
